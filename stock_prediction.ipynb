```python
import pandas as pd 
import datetime as dt
import numpy as np
```

```python
pd.core.common.is_list_like = pd.api.types.is_list_like
from pandas_datareader.data import DataReader
```

```python
#Define dateframe for stock we retrieve
end = dt.datetime.now()
start = end - dt.timedelta(days=5*365)


df = DataReader('AAPL', 'iex', start, end)
df.head(3)
```

```python
df['Log_Ret'] = np.log(df["close"] / df["close"].shift(1))
```

```python
df.tail(2)
```

```python
df['Log_Ret'].quantile(0.25)
```

```python
df['Crash'] = np.where((df['Log_Ret'] < df['Log_Ret'].quantile(0.25)), '1','0')
```

```python
df['Down'] = np.where(((df['Log_Ret'] < df['Log_Ret'].quantile(0.50)) & (df['Log_Ret'] > df['Log_Ret'].quantile(0.25))), '1','0')
```

```python
df['Up'] = np.where(((df['Log_Ret'] > df['Log_Ret'].quantile(0.50)) & (df['Log_Ret'] < df['Log_Ret'].quantile(0.75))), '1','0')
```

```python
df['Jump'] = np.where((df['Log_Ret'] > df['Log_Ret'].quantile(0.75)), '1','0')
```

```python
df.head(2)
```

```python
df.head(20)
```

```python
df["moving_average3"] = df["close"].rolling(3).mean()
```

```python
df["moving_average14"] = df["close"].rolling(14).mean()
```

```python
df["moving_average21"] = df["close"].rolling(21).mean()
```

```python
df.reset_index(inplace=True)
```
